```python
import yfinance as yf
import pandas as pd

# Download FTSE 100 data (10 years)
print("Downloading FTSE 100 data...")

ftse = yf.download("^FTSE", start="2015-01-01", end="2024-12-31")

print("Done!")
print(f"Shape: {ftse.shape}")
print(ftse.head())
```

```python
# Clean the FTSE data
ftse_clean = ftse.copy()

# Flatten column names
ftse_clean.columns = ['Close', 'High', 'Low', 'Open', 'Volume']

# Reset index so Date becomes a column
ftse_clean = ftse_clean.reset_index()

# Rename columns cleanly
ftse_clean.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']

# Check for missing values
print("Missing values:")
print(ftse_clean.isnull().sum())

print(f"\nDate range: {ftse_clean['Date'].min()} to {ftse_clean['Date'].max()}")
print(f"Total trading days: {len(ftse_clean)}")

# Save to our data folder
ftse_clean.to_csv('../data/ftse100_data.csv', index=False)
print("\nFTSE data saved to data folder!")
```
